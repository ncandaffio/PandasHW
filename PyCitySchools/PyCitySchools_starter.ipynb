# PyCity Schools Analysis



* The schools tested seem to have an impact on the students' performance accross all grade groups
* The highest variability on scores seems to be in math. Reading do not differ as much between schools.
* Small and medium sized schools are fairly close in performance, however, larger schools (over 2,000) performed noteably worse

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Add columns for future calculations
school_data_complete['pass_math'] = school_data_complete['math_score'] >= 70
school_data_complete['pass_reading'] = school_data_complete['reading_score'] >= 70
school_data_complete['budget_per_student'] = school_data_complete['budget'] / school_data_complete['size']
school_data_complete.head()
```

```python
district_summary = pd.DataFrame({"Total Schools":[school_data_complete["school_name"].nunique()], 
                                 "Total Students":[school_data_complete["Student ID"].nunique()],
                                "Total Budget":[school_data["budget"].sum()],
                                 "Average Math Score":[school_data_complete["math_score"].mean()],
                                "Average Reading Score":[school_data_complete["reading_score"].mean()], 
                                'Math Pass Rate %':[school_data_complete['pass_math'].mean()],
                                'Reading Pass Rate %':[school_data_complete['pass_reading'].mean()]})
district_summary["Total Pass Rate %"] = (district_summary["Math Pass Rate %"] + district_summary["Reading Pass Rate %"]) / 2

```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_group_mean = school_data_complete.groupby(['school_name']).mean()
school_group_sum = school_data_complete.groupby(['school_name']).sum()
school_group_mean
```

```python

school_strings = pd.DataFrame(school_data_complete[['school_name','type']].drop_duplicates())
school_strings
```

```python
school_list1 = pd.merge(school_group_mean, school_strings, how="left", on=["school_name", "school_name"], )
school_list2 = pd.merge(school_list1, school_group_sum, how="left", on=["school_name", "school_name"], suffixes=('_avg','_sum'))
school_list2['overall_pass_rate'] = (school_list2['pass_math_avg'] + school_list2['pass_reading_avg']) / 2
school_list2.head()
```

```python
top_schools = school_list2.nlargest(5,'overall_pass_rate')
top_schools = top_schools[['school_name', 'type', 'size_avg', 'budget_avg', 'budget_per_student_avg', 'math_score_avg', 'reading_score_avg', 'pass_math_avg', 'pass_reading_avg', 'overall_pass_rate']]
top_schools.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Overall Passing Rate']
top_schools.set_index('School Name', inplace=True)
top_schools['Total School Budget'] = top_schools['Total School Budget'].map("${:,.2f}".format)
top_schools['Per Student Budget'] = top_schools['Per Student Budget'].map("${:,.2f}".format)
top_schools['Total Students'] = top_schools['Total Students'].map("{:,.0f}".format)
top_schools[["Average Math Score", "Average Reading Score"]] = (top_schools[["Average Math Score", "Average Reading Score"]]).applymap("{:,.2f}%".format)
top_schools[["Percent Passing Math", "Percent Passing Reading", "Overall Passing Rate"]] = (top_schools[["Percent Passing Math", "Percent Passing Reading", "Overall Passing Rate"]] * 100).applymap("{:,.2f}%".format)
top_schools
```

```python
top_schools = school_list2.nsmallest(5,'overall_pass_rate')
top_schools = top_schools[['school_name', 'type', 'size_avg', 'budget_avg', 'budget_per_student_avg', 'math_score_avg', 'reading_score_avg', 'pass_math_avg', 'pass_reading_avg', 'overall_pass_rate']]
top_schools.columns = ['School Name', 'School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score', 'Percent Passing Math', 'Percent Passing Reading', 'Overall Passing Rate']
top_schools.set_index('School Name', inplace=True)
top_schools['Total School Budget'] = top_schools['Total School Budget'].map("${:,.2f}".format)
top_schools['Per Student Budget'] = top_schools['Per Student Budget'].map("${:,.2f}".format)
top_schools['Total Students'] = top_schools['Total Students'].map("{:,.0f}".format)
top_schools[["Average Math Score", "Average Reading Score"]] = (top_schools[["Average Math Score", "Average Reading Score"]]).applymap("{:,.2f}%".format)
top_schools[["Percent Passing Math", "Percent Passing Reading", "Overall Passing Rate"]] = (top_schools[["Percent Passing Math", "Percent Passing Reading", "Overall Passing Rate"]] * 100).applymap("{:,.2f}%".format)
top_schools
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
spend_bin_df = school_data_complete
spend_bin_df['spending_bin'] = pd.cut(spend_bin_df['budget_per_student'], spending_bins, labels=group_names)
spend_bin_df = spend_bin_df.groupby('spending_bin').mean()
spend_bin_df = spend_bin_df[['math_score', 'reading_score', 'pass_math', 'pass_reading']]
spend_bin_df['Overall Passing Rate'] = (spend_bin_df['pass_math'] + spend_bin_df['pass_reading']) / 2
spend_bin_df.columns = ['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', 'Overall Passing Rate']
spend_bin_df[['Average Math Score', 'Average Reading Score']] = spend_bin_df[['Average Math Score', 'Average Reading Score']].applymap("{:,.2f}%".format)
spend_bin_df[['% Passing Math', '% Passing Reading', 'Overall Passing Rate']] = (spend_bin_df[['% Passing Math', '% Passing Reading', 'Overall Passing Rate']] * 100).applymap("{:,.2f}%".format)
spend_bin_df
```
